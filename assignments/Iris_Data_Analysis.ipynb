# Problem 4
### Examining the iris dataset itself

- Multivariate dataset introduced by Ronald Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems" as an example of linear discriminant analysis.
- Consists of 50 samples from each of three species of Iris: Iris setosa, Iris virginica, and Iris versicolor, totaling 150 samples.
- Four features measured from each sample:
  - Sepal length (in centimeters)
  - Sepal width (in centimeters)
  - Petal length (in centimeters)
  - Petal width (in centimeters)

- It is widely used in research, education, and benchmarking of classification algorithms due to its simplicity, availability, and well-defined classes.

# Problem 6
### Creating a diagram

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Pie
sr = df["Species"].value_counts()
sr.plot.pie(autopct="%.1f%%");
```

```python
#Using a box plot.

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 4, figsize=(18, 6))  # 1 row, 4 subplots

fig.suptitle('Distribution of data for each label using a box plot')

sns.boxplot(ax=ax[0], x='Species', y='sepal_length', data=df)
sns.boxplot(ax=ax[1], x='Species', y='sepal_width', data=df)
sns.boxplot(ax=ax[2], x='Species', y='petal_length', data=df)
sns.boxplot(ax=ax[3], x='Species', y='petal_width', data=df)

plt.tight_layout()
plt.show()
```

```python
#Violin plot
fig, ax = plt.subplots(2, 2, figsize=(18, 18))

fig.suptitle('Distribution of data for each label using a violin plot')

sns.violinplot(ax=ax[0,0], x='Species', y='sepal_length', data=df)
sns.violinplot(ax=ax[0,1], x='Species', y='sepal_width', data=df)
sns.violinplot(ax=ax[1,0], x='Species', y='petal_length', data=df)
sns.violinplot(ax=ax[1,1], x='Species', y='petal_width', data=df)

sns.set()
```

# Problem 7 
### Confirming the relationship between features

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))

fig.suptitle('Scatter plot')

sns.scatterplot(ax=ax[0,0], data=df, x='sepal_length', y='sepal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,1], data=df, x='sepal_length', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,2], data=df, x='sepal_length', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,0], data=df, x='sepal_width', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,1], data=df, x='sepal_width', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,2], data=df, x='petal_length', y='petal_width', hue='Species', palette="deep")

sns.set()
```
