```python
# Graph showing the percentage of classes
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tg = df["TARGET"].value_counts()
plt.subplots(figsize=(12,8))
pie_target = tg.plot.pie(autopct="%.1f%%")
pie_target.legend(loc=1, labels={'Client with payment difficulties': 0, 'All other cases': 1})
plt.show()
```

# Problem 4
### Data exploration

```python
#percentage of loans
import numpy as np

# Computing the proportions.
loan_outcome = np.array(df['TARGET'])
num_paid_loans = np.count_nonzero(loan_outcome == 0)
num_unpaid_loans = np.count_nonzero(loan_outcome == 1)

# Plotting the results
plt.figure(figsize=(5, 5))
plt.pie(np.array([num_paid_loans, num_unpaid_loans]), 
        autopct='%1.1f%%', 
        startangle=140)
plt.axis('equal')
plt.title('8.1% of loans are paid with some difficulty')
plt.show()
```
