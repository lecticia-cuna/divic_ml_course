# Problem 1
### Creating random numbers

```python
import numpy as np
mean = (-3,0)
covariance = [[1,0.8],[0.8,1]]
np.random.seed(0)

random_ndarray = np.random.multivariate_normal(mean,covariance,size=500)

print(random_ndarray.shape)
print(random_ndarray)
```

# Problem 2
### Visualization by scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

mean = (-3,0)
covariance = [[1,0.8],[0.8,1]]

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatter plot")

x, y = np.random.multivariate_normal(mean, covariance, 500).T
plt.scatter(x, y, color='orange', alpha=0.8)
plt.show()
     
```

# Problem 3
### Visualization by histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("x1")
plt.ylabel("Frequency")
plt.title("Histogram of x1")

x, y = np.random.multivariate_normal(mean, covariance, 500).T
plt.hist(x, color="blue")
plt.xlim(right=4)
plt.show()
     
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("x2")
plt.ylabel("Frequency")
plt.title("Histogram of x2")

x, y = np.random.multivariate_normal(mean, covariance, 500).T
plt.hist(y, color="red")
plt.xlim(left=-5)
plt.show()
```

# Problem 4
### Addition of data

```python
mean2 = (0,-3)
covariance2 = [[1,0.8],[0.8,1]]
np.random.seed(0)
random_ndarray2 = np.random.multivariate_normal(mean2,covariance2,size=500)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

x, y = np.random.multivariate_normal(mean, covariance, 500).T
x2, y2 = np.random.multivariate_normal(mean2, covariance2, 500).T
classes = [0,1]

plt.scatter(x, y, color="orange", alpha=0.5)
plt.scatter(x2, y2, color="green")
plt.legend(labels=classes)
plt.show()
```

# Problem 5
### Data combination

```python
random_ndarray = np.random.multivariate_normal(mean,covariance,size=500)
second_random_ndarray = np.random.multivariate_normal(mean2,covariance2,size=500)

np.random.seed(0)
combined_ndarray=np.concatenate((random_ndarray, second_random_ndarray))

print(combined_ndarray.shape)
print(combined_ndarray)
```
