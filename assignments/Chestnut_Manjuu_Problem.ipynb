# Problem 1

Create a generic program that can be applied to other situations, such as the time it takes for the Tokyo Dome to fill up when you sprinkle a ball with Byvine.

```python
import matplotlib.pyplot as plt


minutes = []
size = []

def time_to_space(obj_size):
    tokyo_dome = 503267  # Size of Tokyo Dome
   
    time = 0  # Initialize time
    while True:
        if obj_size >= tokyo_dome:
            break
        else:
            obj_size *= 2
            time += 5
            minutes.append(time)
            size.append(obj_size)
    return time  # Return the time it took to fill up the Tokyo Dome

time = time_to_space(100000)

# Plotting
plt.plot(minutes, size, 'b-o', linewidth=3, markersize=15, markeredgewidth=5, markeredgecolor='green', label='Time taken to fill up Tokyo Dome')
plt.xlabel("Time in minutes")
plt.ylabel("Size of the object in cubic meters")
plt.title("Time to Cover the Tokyo Dome")
plt.show()

print("It takes {} minutes".format(time))
```
